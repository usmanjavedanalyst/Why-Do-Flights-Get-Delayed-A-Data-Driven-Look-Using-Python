```python
import pandas as pd

df = pd.read_csv("2017.csv")
df.head()
df.info()
df.shape
```

```python
df = df[df["CANCELLED"] == 0]
df = df.dropna(subset=["DEP_DELAY", "ARR_DELAY"])
df.shape
```

```python
avg_dep_delay = df["DEP_DELAY"].mean()
avg_dep_delay
```

```python
avg_arr_delay = df["ARR_DELAY"].mean()
avg_arr_delay
```

```python
delayed_flights = df[df["DEP_DELAY"] > 0]

delayed_percentage = (len(delayed_flights) / len(df)) * 100
delayed_percentage
```

```python
airline_delays = df.groupby("OP_CARRIER")["DEP_DELAY"].mean()
airline_delays = airline_delays.sort_values(ascending=False)

airline_delays.head(10)
```

```python
airline_delays = df.groupby("OP_CARRIER")["DEP_DELAY"].mean()
airline_delays = airline_delays.sort_values(ascending=True)

airline_delays.head(10)
```

```python
airline_fl_delays = df.groupby("OP_CARRIER_FL_NUM")["DEP_DELAY"].mean()
airline_fl_delays = airline_fl_delays.sort_values(ascending=False)

airline_fl_delays.head(10)
```

```python
airport_delays = df.groupby("ORIGIN")["DEP_DELAY"].mean()

airport_delays = airport_delays.sort_values(ascending=False)

airport_delays.head(10)
```

```python
airport_delays = df.groupby("ORIGIN")["DEP_DELAY"].mean()

airport_delays = airport_delays.sort_values(ascending=True)

airport_delays.head(10)
```

```python
df["FL_DATE"] = pd.to_datetime(df["FL_DATE"])

df["MONTH"] = df["FL_DATE"].dt.month

monthly_delay = df.groupby("MONTH")["DEP_DELAY"].mean()
monthly_delay
```

```python
df["DISTANCE_BIN"] = pd.cut(
    df["DISTANCE"],
    bins=[0, 500, 1000, 1500, 2000, 3000]
)

distance_delay = df.groupby("DISTANCE_BIN", observed=True)["DEP_DELAY"].mean()
distance_delay
```
